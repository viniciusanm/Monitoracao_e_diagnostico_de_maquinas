```python
df_acc_1 = pd.read_csv('D:/DR_train/Monitoracao_diagnostico_maquinas/Python/df_acc_1.csv', dtype={'Daily_passing' : 'int16'})
df_acc_2 = pd.read_csv('D:/DR_train/Monitoracao_diagnostico_maquinas/Python/df_acc_2.csv', dtype={'Daily_passing' : 'int16'})
df_acc_3 = pd.read_csv('D:/DR_train/Monitoracao_diagnostico_maquinas/Python/df_acc_3.csv', dtype={'Daily_passing' : 'int16'})
df_acc_4 = pd.read_csv('D:/DR_train/Monitoracao_diagnostico_maquinas/Python/df_acc_4.csv', dtype={'Daily_passing' : 'int16'})
df_acc_5 = pd.read_csv('D:/DR_train/Monitoracao_diagnostico_maquinas/Python/df_acc_5.csv', dtype={'Daily_passing' : 'int16'})
```

```python
x_1 = df_acc_1['Track Position']
y_1 = df_acc_1['Daily_passing']
z_1 = df_acc_1['Triaxial_y']

x_2 = df_acc_2['Track Position']
y_2 = df_acc_2['Daily_passing']
z_2 = df_acc_2['Triaxial_y']

x_1 = df_acc_1['Track Position']
y_1 = df_acc_1['Daily_passing']
z_1 = df_acc_1['Triaxial_y']

x_2 = df_acc_2['Track Position']
y_2 = df_acc_2['Daily_passing']
z_2 = df_acc_2['Triaxial_y']

x_3 = df_acc_3['Track Position']
y_3 = df_acc_3['Daily_passing']
z_3 = df_acc_3['Triaxial_y']

x_4 = df_acc_4['Track Position']
y_4 = df_acc_4['Daily_passing']
z_4 = df_acc_4['Triaxial_y']

x_5 = df_acc_5['Track Position']
y_5 = df_acc_5['Daily_passing']
z_5 = df_acc_5['Triaxial_y']
```

```python
jet=plt.get_cmap('jet')

plt.scatter(x_1, y_1, c=z_1, cmap=jet)

plt.colorbar()
plt.show()
```

```python
jet=plt.get_cmap('jet')

plt.scatter(x_1, y_1, c=z_1, cmap=jet)
plt.scatter(x_2, y_2, c=z_2, cmap=jet)
plt.scatter(x_3, y_3, c=z_3, cmap=jet)
plt.scatter(x_4, y_4, c=z_4, cmap=jet)
plt.scatter(x_5, y_5, c=z_5, cmap=jet)

plt.grid(True)

cbar = plt.colorbar()
cbar.set_label('Vehicle Acceleration (g)')

plt.xlabel('Track Position (m)')
plt.ylabel('Daily Passing')

plt.show()

plt.savefig('acc_daily_pass.png')
```
