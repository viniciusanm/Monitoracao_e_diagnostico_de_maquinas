# Alinhando os DataFrames

```python
import numpy as np
from scipy.spatial import cKDTree
```

```python
#Create the datasets
df_gps_gt2_al = df_gps_gt_2[['Longitude', 'Latitude']]
df_gps_al = df_gps[['Longitude', 'Latitude']]
```

```python
tree = cKDTree(df_gps_gt2_al[['Longitude', 'Latitude']].values)
dists, indices = tree.query(df_gps_al[['Longitude', 'Latitude']].values, k=1)
fts = [c for c in df_gps_gt2_al.columns]
for c in fts:
    df_gps_al[c] = df_gps_gt2_al[c].values[indices]

df_gps_final = df_gps_al
```

```python
plt.scatter(df_gps_final['Longitude'], df_gps_final['Latitude'], label = 'Região 2 Alinhada')
plt.scatter(df_gps_gt_2['Longitude'], df_gps_gt_2['Latitude'], label = 'G.T. Região 2')
plt.legend()
plt.title('Após Alinhamento')

plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.savefig('comparacao_alinhado.png')
```
